```python
import json
import requests
import pandas as pd
```

```python
data = requests.get('https://statsapi.mlb.com/api/v1/schedule/games/?sportId=1&startDate=2015-01-01&endDate=2022-12-31')
```

```python
data = data.text
```

```python
json_data = json.loads(data)
```

```python

json_data = json.loads(data)
dates = json_data['dates']

for date in dates:
    game_date = date['date']
    games = date['games']
    for game in games:
        season = game['season']
        game_type = game['gameType']
        home_team = game['teams']['home']['team']['name']
        away_team = game['teams']['away']['team']['name']
        home_score = game['teams']['home'].get('score', 'Unknown')
        away_score = game['teams']['away'].get('score', 'Unknown')
        home_is_winner = game['teams']['home'].get('isWinner','Unknown')
        away_is_winner = game['teams']['away'].get('isWinner','Unknown')

        print(f"Date: {game_date}, Season: {season}, Home Team: {home_team}, Away Team: {away_team}, Home Score: {home_score}, Away Score: {away_score}, Home isWinner: {home_is_winner}, Away isWinner: {away_is_winner}, Game Type: {game_type}")
```
